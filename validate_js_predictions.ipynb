```python
import pandas as pd
```

```python
test_predictions = pd.read_csv("data/test_predictions.csv").set_index("filename")
js_predictions = pd.read_csv("data/node_predictions.csv").set_index("filename")
```

```python
predictions = (
    test_predictions
    .join(js_predictions, how = "outer", lsuffix = "_test", rsuffix = "_js")
    .rename(columns = { "prediction_test": "test", "prediction_js": "js" })
)
```

```python
predictions.head(5)
```

```python
prediction_match_ratio = predictions.apply(lambda r: r.test == r.js, axis=1).mean()
prediction_match_ratio
```

```python
assert prediction_match_ratio > .99
```
